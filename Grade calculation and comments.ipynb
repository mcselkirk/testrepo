```python
#import important modules and read .csv into a dataframe
#don't forget to clean up data by copying the Speaker into all rows to avoid missing invisible spaces that will then give 2 rows if not remedied
import pandas as pd

df = pd.read_csv('2022statpres1_aliases.csv')

df.head()
```

```python
# Give size of dataset
df.shape
```

```python
# Calculate mean of all numeric columns in dataframe and make a new dataframe
df2 = df.groupby('Speaker').mean()
df2
```

```python
# Sort output by period, then speaker

df3 = df2.sort_values(by=['Period', 'Speaker'])
df3
```

```python
# Get rid of numeric column I don't want in there
df4 = df3.drop(['Period'], axis=1)
df4
```

```python
# Get total for each row
df5 = pd.concat([df4,pd.DataFrame(df4.sum(axis=1),columns=['Total'])],axis=1)
df5
```

```python
# Make a new column calculating score and add to dataframe using a simple function (lambda)
df5['Score'] = df5.apply(lambda row: row.Total/25*100 , axis=1)
pd.set_option('display.max_rows', None)
df5
```
